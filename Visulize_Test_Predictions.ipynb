```python
import pandas as pd
```

```python
import numpy as np
from seaborn import heatmap
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
air_visit_data = pd.read_csv('air_visit_data.csv')
```

```python
air_visit_data['visit_date'] = pd.to_datetime(air_visit_data['visit_date'],format= '%Y-%m-%d')
```

```python
air_visit_data.columns = ['air_store_id', 'visit_date_2016', 'visitors']
```

```python
xgb_res_474 = pd.read_csv('1.csv')
xgb_res_474['air_store_id'] = xgb_res_474['id'].apply(lambda x: x.split('_')[0]+str('_') +x.split('_')[1])
xgb_res_474['visit_date'] = xgb_res_474['id'].apply(lambda x: x.split('_')[2])
xgb_res_474['visit_date'] = pd.to_datetime(xgb_res_474['visit_date'],format= '%Y-%m-%d %H:%M:%S')
```

```python
N = 390
air_store_ids = np.unique(xgb_res_474.air_store_id[:N])
```

```python
air_store_ids
```

```python
# 1 year difference for the same date
xgb_res_474['visit_date_2016'] = xgb_res_474['visit_date'].apply(lambda x: x - pd.DateOffset(years=1))
```

```python
df = pd.merge(xgb_res_474, air_visit_data, how = 'left',
                  left_on = ['air_store_id','visit_date_2016'],
                  right_on = ['air_store_id','visit_date_2016'])
```

```python
k = df[df['air_store_id'].isin(air_store_ids)]
```

```python
k.replace(np.nan,0,inplace = True)
```

```python
N = 390
plt.figure(figsize=(16,8))
plt.plot(range(N), k.visitors_x[:N], label = 'Xgb')
for i in range(0, 391, 39):
    plt.axvline(x=i)
plt.legend()
```
